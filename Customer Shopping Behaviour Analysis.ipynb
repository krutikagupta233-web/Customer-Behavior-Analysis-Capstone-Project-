```python
# Here as we saw that there are 37 null values in review rating so we are relacing it by mean of review rating as per their categories
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()  # Now here we can see there are no null values in Review Rating Column
```

```python
# Now as we can see the column names have uppercase,lowercase letters and space between it so it will be difficult to remember in sql.
# So to make it consistent we would make our column names in "Snake Casing".
# snake casing means all letters in lowercase and instesd of space use underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")

# Here Renaming column "purchase_amount_(usd)" to "purchase_amount"
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
# Now we can see all column names are in snake casing
df.columns
```

#### Feature Engineering

```python
# Creating a new column name as age_group 
labels=["Young_Adult","Adult","Middle_Aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
# Create Column "purchase_frequency_days"
# We are creating this column bcz the existing column which is "frequency_of_purchases" in text but nuerical columns are more prefferable for analysis 
# so we have created a numerical column of it i.e. "purchase_frequency_days"

frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Quarterly":90,
    "Monthly":30,
    "Every 3 Months":90,
    "Bi-Weekly":14,
    "Annually":365
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)

# Let's Check the results
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```

```python
# Now lets look at two columns which are "discount_applied" and "promo_code_used"
df[["discount_applied","promo_code_used"]].head(10)
```

```python
# Now let's evaluate that actually both columns are containing equal values or not
(df["discount_applied"] == df["promo_code_used"]).all() #Here all means match with all rows 
```

```python
# Now as we can see both columns are containing same values so let's delete the column "promo_code_used"
df=df.drop("promo_code_used",axis=1)
```
